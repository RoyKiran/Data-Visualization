```python
# Lets get some random dataframe
df1 = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())   #randn is for normal distribution with 100 rows & 4 columns named A, B, C & D
df1.head()
```

```python
# Pandas plot
df1.plot()
```

```python
# to plot more interactive graph instead of plot() we use iplot()
df1.iplot()
```

## Using Cufflinks and iplot()

* scatter
* bar
* box
* spread
* ratio
* heatmap
* surface
* histogram
* bubble

### Scatter Plot

```python
# Interactive Scatter Plot
#df1.iplot(kind='scatter', x='A', y='B')  #Points are connected using lines by default
#df1.iplot(kind='scatter', x='A', y='B', mode='markers')  #Points are connected using dots
df1.iplot(kind='scatter', x='A', y='B', mode='markers', size=25)   #Setting dots/markers size
```

### Bar Plot

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
# If you plot all the fields in dataframe
df1.iplot(kind='bar')
```

```python
# Plot aggregate function of## Boxplots a dataframe
#df1.count().iplot(kind='bar')
df1.sum().iplot(kind='bar')
```

### Box Plot

```python
# Box Plot
df1.iplot(kind='box')
```

### 3D Surface Plot

```python
# Surface plot
#df3.iplot(kind='surface')
df3.iplot(kind='surface', colorscale='rdylbu')  #rd=red, yl=yellow, bu=blue
```

### Histogram plot

```python
# Histogram plot
#df1['A'].iplot(kind='hist')
#df1.iplot(kind='hist')
df1['A'].iplot(kind='hist', bins=50)
```

### Spread Plot

```python
# Spread plot
#df1[['A','B']]
df1[['A','B']].iplot(kind='spread')
```

### Bubble Plot

```python
# Bubble plot
df1.iplot(kind='bubble', x='A', y='C', size='B')
```

### scatter_matrix
Similar to sns.pairplot()

```python
# Scatter Matrix
df1.scatter_matrix()
```
